```python
def binarySearchIndexAndComparisons(L, k):
    left = 0
    right = len(L)-1
    count = 0
    while left <= right:
        mid = (right + left) // 2
        if L[mid] == k:
            count += 1
            return (True, count)
        else:
            count += 1
            if L[mid] > k:
                right = mid-1
            else:
                left = mid + 1
    return (False, count)
```

```python
#python
def mergeInPlace(A, B):
    m = len(A)
    n = len(B)
    if (m < 1 or n < 1):
        return

    # Find the smaller list of A and B.
    for i in range(0, m):
        # A and B are already sorted. B[0] will always be least in B,
        # as we will maintain its sortedness .
        if (A[i] > B[0]):
            A.swap(i, B, 0)

        # move `B[0]` to its correct position in B to maintain the sortedness of B
        j = 0
        while(j < n - 1 and B[j] > B[j + 1]):
            B.swap(j+1, B, j)
            j += 1
    return A, B
```

```python
mergeInPlace([2,4,6,9,13,15], [1, 3, 5, 10])
```

```python
def sortInRange(L, r):
    # Create a dictionary with r keys for each integer in range r, initialize
    # every value to 0
    countDict = dict.fromkeys(range(r), 0)
    # Iterate over the array and count each integer in the list
    for num in L:
        countDict[num] += 1
    index=0
    for key in countDict.keys():
        for i in range(countDict[key]):
            L[index] = key
            index += 1
    return L
```

```python
def fun(n):
    k=0
    for i in range(n//2, n):
        j = 2
        while j <= n:
            k+= n//2
            j*=2
    return k
```
